```python
movies['release_date'] = movies['release_date'].astype(str)
```

```python
def ExtractYear(val):
    tokens = val.split('-')
    return tokens[0]
```

```python
movies['release_year'] = movies['release_date'].apply(ExtractYear)
```

## Return on Investment

```python
money = movies[movies.revenue != 0]
```

```python
money = money[money.budget != 0]
```

```python
money['profit'] = money['revenue'] - money['budget']
```

```python
money['profit'].describe()
```

```python
profit = money['profit']

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.boxplot(profit)
plt.title('Revenue Boxplot')

plt.subplot(1, 2, 2)
plt.hist(profit)
plt.title('Revenue Histogram')
plt.axvline(profit.mean(), color='c', linestyle='solid', linewidth=2, label="Mean")
plt.legend(loc='upper right') 

plt.show()
```

```python
money['profitPercent'] = money['profit'] / money['budget']
```

```python
money['profitPercent'].describe()
```

```python
percent = money['profitPercent']

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.boxplot(percent)
plt.title('Revenue Boxplot')

plt.subplot(1, 2, 2)
plt.hist(percent, bins=20)
plt.title('Revenue Histogram')
plt.axvline(percent.mean(), color='c', linestyle='solid', linewidth=2, label="Mean")
plt.legend(loc='upper right') 

plt.show()
```
