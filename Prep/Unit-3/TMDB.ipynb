```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
%matplotlib inline
```

```python
movies = pd.read_csv('~/Desktop/Data_Science/tmdb/tmdb_5000_movies.csv')
```

```python
movies.head()
```

```python
movies.describe()
```

```python
movies['release_date'] = movies['release_date'].astype(str)
```

```python
def ExtractYear(val):
    tokens = val.split('-')
    return tokens[0]
```

```python
movies['release_year'] = movies['release_date'].apply(ExtractYear)
```

```python
movies.head()
```

```python
movies['release_year'] = pd.to_numeric(movies['release_year'], errors='coerce')
```

```python
movies['release_year'].describe()
```

```python
plt.boxplot(movies['release_year'])
plt.show()
```

```python
plt.hist(movies['release_year'])
plt.show()
```

```python
movies['release_year'] = movies['release_year'].astype(np.float)
```

```python
plt.figure(figsize=(30, 30))

movies[['release_year', 'revenue']].boxplot(by='release_year', figsize=(10,20))
```

## Year

```python
year_info = movies.groupby('release_year').aggregate(np.mean)
```

```python
year_info.head()
```

```python
plt.figure(figsize=(15, 5))

plt.scatter(x=year_info['release_year'], y=year_info['budget'])
plt.xticks(rotation = 90)
plt.title('Average Budget by Year of Release')
plt.xlabel('Year of Release')
plt.ylabel('Budget by 100 million')

plt.show()
```

```python
plt.figure(figsize=(15, 5))

year_info.plot(y=["budget", "revenue"])
```

## Language

```python
# Process data to create counts by original language

languages = movies.original_language.tolist()
languages = [x for x in languages if str(x) != 'nan']
language = pd.DataFrame(languages, columns=['language'])

# Create numeric counts
lang_counts = pd.DataFrame(language.language.value_counts())
other = int(lang_counts[10:].sum())
lang_counts = lang_counts[:10]
lang_counts.loc[11] = other
lang_counts = lang_counts.rename(index={11: 'Other'})

# Generate Pie Chart
plt.figure(figsize=(10, 5))
plt.pie(lang_counts.language)
plt.axis('equal')
plt.title('Original Language')
plt.legend(lang_counts.index)
```
