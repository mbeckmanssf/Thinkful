```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
%matplotlib inline
```

```python
movies = pd.read_csv('~/Desktop/Data_Science/tmdb/tmdb_5000_movies.csv')
```

```python
movies.head()
```

```python
movies.describe()
```

```python
movies['release_date'] = movies['release_date'].astype(str)
```

```python
def ExtractYear(val):
    tokens = val.split('-')
    return tokens[0]
```

```python
movies['release_year'] = movies['release_date'].apply(ExtractYear)
```

```python
movies.head()
```

```python
pd.to_numeric(movies['release_year'], errors='coerce')
```

```python
plt.boxplot(movies['release_year'])
plt.show()
```

```python
movies['release_year'] = movies['release_year'].astype(np.float)
```

## Budget

```python
#budget

budget = movies['budget']

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.boxplot(budget)
plt.title('Budget Boxplot')

plt.subplot(1, 2, 2)
plt.hist(budget)
plt.title('Budget Histogram')
plt.axvline(budget.mean(), color='c', linestyle='solid', linewidth=2, label="Mean")
plt.legend(loc='upper right') 

plt.show()
```

## Revenue

```python
#revenue

revenue = movies['revenue']

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.boxplot(revenue)
plt.title('Revenue Boxplot')

plt.subplot(1, 2, 2)
plt.hist(revenue)
plt.title('Revenue Histogram')
plt.axvline(revenue.mean(), color='c', linestyle='solid', linewidth=2, label="Mean")
plt.legend(loc='upper right') 

plt.show()
```

## Budget and Revenue

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
plt.scatter(x=movies['budget'], y=movies['revenue'])
plt.title('Revenue by Budget')
plt.xlabel('Budget by 100 million')
plt.ylabel('Revenue by billion')

plt.subplot(1, 2, 2)
plt.scatter(x=movies['revenue'], y=movies['budget'])
plt.title('Budget by Revenue')
plt.xlabel('Revenue by billion')
plt.ylabel('Budget by 100 million')

plt.show()
```

## Vote Average

```python
#vote average

average = movies['vote_average']

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.boxplot(average)
plt.title('Vote Average Boxplot')

plt.subplot(1, 2, 2)
plt.hist(average)
plt.title('Vote Average Histogram')
plt.axvline(average.mean(), color='c', linestyle='solid', linewidth=2, label="Mean")
plt.legend(loc='upper right') 

plt.show()
```
